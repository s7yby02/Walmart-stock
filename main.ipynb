```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Download Walmart stock data
ticker = "WMT"
data = yf.download(ticker, start="2015-01-01", end="2025-01-20")

# Display the first few rows of the data
data.head()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Close'], label='Walmart Closing Price')
plt.title('Walmart Stock Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Volume'], label='Trading Volume', color='orange')
plt.title('Walmart Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.show()
```
